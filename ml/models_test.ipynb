```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
import tensorflow as tf
import urllib
from dataclasses import dataclass
import tensorflowjs as tfjs
import json
import keras
```

```python
urllib.request.urlretrieve(
    "https://github.com/ur-whitelab/peptide-dashboard/raw/master/ml/data/hemo-positive.npz",
    "hemo-positive.npz",
)
urllib.request.urlretrieve(
    "https://github.com/ur-whitelab/peptide-dashboard/raw/master/ml/data/hemo-negative.npz",
    "hemo-negative.npz",
)
with np.load("hemo-positive.npz") as r:
    hemo_pos_data = r[list(r.keys())[0]]
with np.load("hemo-negative.npz") as r:
    hemo_neg_data = r[list(r.keys())[0]]

# create labels and stich it all into one
# tensor
hemo_labels = np.concatenate(
    (
        np.ones((hemo_pos_data.shape[0], 1), dtype=hemo_pos_data.dtype),
        np.zeros((hemo_neg_data.shape[0], 1), dtype=hemo_pos_data.dtype),
    ),
    axis=0,
)

hemo_features = np.concatenate((hemo_pos_data, hemo_neg_data), axis=0)
```
